```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
df = df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Cabin','Embarked', 'Ticket'], axis = 'columns')
df
```

```python
input = df.drop(['Survived'], axis = 'columns')
input.head()
```

```python
target = df['Survived']
target
```

```python
#Covert string to numerical value in Sex
dummies = pd.get_dummies(input['Sex'], dtype = 'int')
dummies.head()
```

```python
input = input.drop(['Sex'], axis = 'columns')
input
```

```python
input = pd.concat([input, dummies], axis = 'columns')
input
```

```python
input.head(10)
```

```python
input.Age
```

```python
input.Age = input.Age.fillna(input.Age.mean())
input.head(10)
```

```python
X = input
X.head()
```

```python
y = target
y.head()
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```
